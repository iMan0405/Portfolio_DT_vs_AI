```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error,r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/EURUSD=X .csv')
```

```python
df.head()
```

```python
viz = df.copy
```

#Data Preparation

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe().T
```

```python
train, test = train_test_split(df, test_size = 0.2)
```

```python
test_pred = test.copy()
```

```python
train.head(10)
```

```python
test.head(10)
```

```python
x_train = train[['Open', 'Low', 'Volume', 'Close']].values
x_test = test[['Open', 'Low', 'Volume', 'Close']].values
```

```python
y_train = train['High'].values
y_test = test['High'].values
```

#Linear Regression

```python
model_lnr = LinearRegression()
model_lnr.fit(x_train, y_train)
```

```python
y_pred = model_lnr.predict(x_test)
```

```python
result = model_lnr.predict([[1.090001, 1.089752, 1.090037, 1.090037]])
print(result)
```

#Model Evaluation

```python
print("MSE",round(mean_squared_error(y_test,y_pred), 3))
print("RMSE",round(np.sqrt(mean_squared_error(y_test,y_pred)), 3))
print("MAE",round(mean_absolute_error(y_test,y_pred), 3))
print("MAPE",round(mean_absolute_percentage_error(y_test,y_pred), 3))
print("R2 Score : ", round(r2_score(y_test,y_pred), 3))
```
